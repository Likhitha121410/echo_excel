```python
import pandas as pd
data=pd.read_csv("/content/exce2.csv")
print(data)
print("------------")

for i in range (len(data)):
  print(data.iloc[i])
  print("----------")
  print("------------------")
a=data[data["internet speed"]>=10]
print(a)
print("__________")
print("__________")
b=a[a["payment failuer"]<=3]
print(b)
print("_____________")
print("_____________")
c=b[b["del_charges"]<80]
print(c)
```
