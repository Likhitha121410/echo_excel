```python
import statistics
member_names=["likki","junnu","nanda","mihin","yuvan"]
ages=[21,6,20,6,8]
workout_hours=[2,1.5,6,4,3]
cal_burns=[200,500,450,670,500]
print("-----Fitness Center Dataset-------")
for i in range(len(member_names)):
  print(member_names[i],
      ages[i],
      workout_hours[i],
      cal_burns[i])
  #avg
  avg_age=sum(ages)/len(ages)
  print("average age=",avg_age)
# MAX
highest_cal=max(cal_burns)
print("Highest Cal burn=",highest_cal)
# min
lowest_workout=min(workout_hours)
print("lower workout hours=",lowest_workout)
# median
median_cal=statistics.median(cal_burns)
print("median calories=",median_cal)
#std
std_dev=statistics.stdev(cal_burns)
print("std=",std_dev)
# Active index
index = cal_burns.index(highest_cal)
print("most active member=",member_names[index])
#visulazition
import matplotlib.pyplot as plt
plt.bar(member_names,cal_burns)
plt.xlabel("Members")
plt.ylabel("Calories Burned")
plt.title("Fitness Analysis")
plt.show()
```
